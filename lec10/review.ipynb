# Gaussian Filter

```python
import matplotlib.pyplot as plt
from skimage import io


img = io.imread('data/scis1.jpeg')
plt.imshow(img)
plt.show()
```

# Task 1. Create a 1d Gaussian Filter

```python
from scipy.signal.windows import gaussian

g1d = gaussian(9, std=3).reshape(9, 1)
plt.plot(g1d)
plt.show()
```

# Task 2. Create a 2d Gaussian Filter

```python
import numpy as np

g2d = np.outer(g1d, g1d)
plt.imshow(g2d, interpolation='none')
plt.show()
```

```python
g2d
```

# Task 3. Apply the Gaussian filter to an image

```python
from skimage.color import rgb2gray
img = rgb2gray(img)
```

```python
from scipy.ndimage import convolve

img2 = convolve(img, g2d)

fig, axs = plt.subplots(1, 2)
fig.set_dpi(150)

axs[0].imshow(img, cmap='gray')
axs[1].imshow(img2, cmap='gray')

plt.show()
```
