# Transformations

## Rotation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.pi / 4
R = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta), np.cos(theta), 0],
    [0, 0, 1]
])
R
```

```python
a = np.array([1, 0, 1])
b = np.dot(R, a)
print(f"{a}")
print(f"{b}")
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, aspect=1)


circle = plt.Circle((0, 0), 1., color='k', fill=False)
ax.add_patch(circle)

points = np.array([a, b])
ax.scatter(points[:,0], points[:,1], 10**2, color='orange')

ax.set_xlim((-1.7, 1.7))
ax.set_ylim((-1.7, 1.7))


plt.show()
```

```python
RR = np.dot(R, R)
c = np.dot(RR, a)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, aspect=1)


circle = plt.Circle((0, 0), 1., color='k', fill=False)
ax.add_patch(circle)

points = np.array([a, b, c])
ax.scatter(points[:,0], points[:,1], 10**2, color='orange')

ax.set_xlim((-1.7, 1.7))
ax.set_ylim((-1.7, 1.7))


plt.show()
```

## Shear

```python
U = np.array([
    [1, 0.5, 0],
    [0, 1, 0],
    [0, 0, 1]
])
U
```

```python
a = np.array([0, 1, 1])
a2 = np.dot(U, a)
b = np.array([0, 0.5, 1])
b2 = np.dot(U, b)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, aspect=1)


circle = plt.Circle((0, 0), 1., color='k', fill=False)
ax.add_patch(circle)

points = np.array([a, b])
ax.scatter(points[:,0], points[:,1], 10**2, color='gray')
points = np.array([a2, b2])
ax.scatter(points[:,0], points[:,1], 10**2, color='orange')

ax.set_xlim((-1.7, 1.7))
ax.set_ylim((-1.7, 1.7))


plt.show()
```

## Affine

### Task 1

```python
R = np.array([
    [np.cos(np.pi/2), -np.sin(np.pi/2), 0],
    [np.sin(np.pi/2), np.cos(np.pi/2), 0],
    [0, 0, 1]
])
R
```

```python
S = np.array([
    [2, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])
S
```

```python
T = np.array([
    [1, 0, 1],
    [0, 1, 0],
    [0, 0, 1]
])
T
```

```python
U = np.array([
    [1, 2, 0],
    [0, 1, 0],
    [0, 0, 1]
])
U
```

```python
a = np.array([0, 0, 1]).T
a
```

```python
a2 = np.dot(T, a)
ST = np.dot(S, T)
a3 = np.dot(ST, a)
RST = np.dot(R, ST)
a4 = np.dot(RST, a)
URST = np.dot(U, RST)
a5 = np.dot(URST, a)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, aspect=1)


circle = plt.Circle((0, 0), 1., color='k', fill=False)
ax.add_patch(circle)

points = np.array([a, a2, a3, a4, a5])
ax.scatter(points[:,0], points[:,1], 10**2, color='orange')

ax.set_xlim((-5, 5))
ax.set_ylim((-5, 5))


plt.show()
```

### Task 3

```python
x1 = np.array([1, 0, 1])
x2 = np.array([0, 1, 1])
x3 = np.array([1, 1, 1])

A = URST
```

```python
x1_ = np.dot(A, x1.T)
x2_ = np.dot(A, x2)
x3_ = np.dot(A, x3)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, aspect=1)


circle = plt.Circle((0, 0), 1., color='k', fill=False)
ax.add_patch(circle)

points = np.array([x1, x2, x3])
ax.scatter(points[:,0], points[:,1], 10**2, color='gray')
points = np.array([x1_, x2_, x3_])
ax.scatter(points[:,0], points[:,1], 10**2, color='orange')

ax.set_xlim((-10, 10))
ax.set_ylim((-10, 10))


plt.show()
```

# Signal and Filters

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500, endpoint=False)
x1 = np.sin(2 * np.pi * t * 5)
plt.plot(t, x1)
plt.ylim(-2, 2)
plt.show()
```

```python
# from scipy import signal
# x = signal.square(2 * np.pi * 5 * t)

x2 = np.sign(x1)
plt.plot(t, x2)
plt.ylim(-2, 2)
plt.show()
```

```python
diff = np.array([1, 0, -1])
dsig = np.convolve(x2, diff)
# plt.stem(dsig);
plt.plot(dsig)
plt.show()
```
